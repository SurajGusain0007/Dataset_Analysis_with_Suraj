#Exploratory Data Analysis

```python
#import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

**Reading a dataset**

```python

data=pd.read_csv("zomato.csv",encoding='latin-1')
data.head(5)
```

**Show the columns**

```python

data.columns
```

```python
#Information of the dataset
data.info()
```

```python
#Stastics Summmary of the dataset
data.describe()
```

```python
#Finding Missing values
data.isnull().sum()
```

```python
#Using for loop to find columns which contain null values
[features for features in data.columns if data[features].isnull().sum()>0]
```

```python
#Visulization of heatmap to check null values
plt.figure(figsize=(16,9))
sns.heatmap(data.isnull(),cmap='viridis',yticklabels=False)
plt.show()
```

```python
# Reading the dataset Country-Code
df=pd.read_excel("Country-Code.xlsx")
df.head(5)
```

```python
#Merge the two table
final_data=pd.merge(data,df,on='Country Code',how='left')
final_data
```

```python
#To check datatypes
final_data.dtypes
```

```python
# Count no countries in Dataset
final_data.Country.value_counts()
```

```python
#Index 
country=final_data.Country.value_counts().index
country
```

```python
country_val=final_data.Country.value_counts().values
country_val
```

```python
#Top 3 countries uses zomato
plt.figure(figsize=(16,9))
plt.pie(country_val[:3],labels=country[:3],autopct="%1.2f%%")
plt.show()
```

```python
#Observation:Zomato maximum records or transaction are from india then USA and after United Kingdom
```

```python
#Numerical variables
ratings=final_data.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:"Rating count"})
ratings
```

```python
#Observation:When rating is 4.5 to 4.9--->Excellent
#Observation:When rating is 4.0 to 4.94-->Very good
#Observation:When rating is 3.5 to 3.9--->Good
#Observation:When rating is 2.5 to 3.4--->Average
#Observation:When rating is 1.8 to 2.4--->Poor
#Observation:When rating is 0.0--->No rating


```

```python
ratings.head(2)
```

```python
#Display the visulization of pie chart 
plt.figure(figsize=(16,9))
sns.barplot(x='Aggregate rating',y='Rating count',data=ratings,hue='Rating color',palette=['white','red','yellow','green','pink'])
plt.show()
```

```python
#Observations:#Not rated count is high
#Maximun no of ratings 2.4 to 3.4
```

```python
#Display the visulization of countplot
plt.figure(figsize=(16,9))
sns.countplot(x='Rating text',data=ratings)
plt.show()
```

```python
#Display the visulization of countplot to count Rating color
plt.figure(figsize=(16,9))
sns.countplot(x='Rating color',data=ratings)
plt.show()
```
