**8.Rename Month Names to English**

```python
data['month_new']=data['month'].map({'Janeiro':'jan',
                                     'Fevereiro':'feb',
                                     'Margo':'march',
                                     'Abril':'april',
                                     'Maio':'may',
                                     'Junho':'jun',
                                     'Julho':'july',
                                     'Agosto':'august',
                                     'Setembro':'sep',
                                     'Outubro':'October',
                                     'Novembro':'nov',
                                     'Dezembro':'December'})
data
```

**9.Total no of files registered**

```python
data.shape
```

**10.In which month maximum no of forest fires were reported**

```python
data1=data.groupby('month_new')['number'].sum().reset_index()
data1
```

```python
plt.figure(figsize=(16,5))
sns.barplot(x='month_new',y='number',data=data1,hue="month_new",dodge=False,)
plt.title("Maximum no of forest files registred")
plt.legend(bbox_to_anchor=(1.05,1),loc=2)
```

**11.In which year maximum no of files were registered**

```python
data2=data.groupby('year')['number'].sum().reset_index()
data2
```

```python
plt.figure(figsize=(16,5))
sns.barplot(x='year',y='number',data=data2,hue='year',dodge=False)
plt.legend(bbox_to_anchor=(1.05,1),loc=2)
```

**12.In which state maximum no of files were reported**

```python
data3=data.groupby("state")['number'].sum().reset_index()
data3
```

```python
plt.figure(figsize=(16,9))
sns.barplot(x='state',y='number',data=data3,dodge=False,hue='state')
plt.legend(bbox_to_anchor=(1.05,1),loc=2)
plt.xticks(rotation=75)
plt.show
```

**12.Find the total number of fires were registered in amazon**

```python
data[data['state']=='Amazonas']['number'].sum()
```

**13.Display no of fires were reported in amazonas(year wise)**

```python
data4=data[data['state']=='Amazonas']
data5=data4.groupby('year')['number'].sum().reset_index()
data5
```

```python
plt.figure(figsize=(16,9))
sns.barplot(x='year',y='number',data=data5,dodge=False,hue='year')
plt.legend(bbox_to_anchor=(1.05,1),loc=2)
plt.xticks(rotation=75)
plt.show
```

**Find total no of fires were reported in 2015 and visualize data in each month**

```python
data6=data[data['year']==2015]
data6
```

```python
data6.groupby('month_new')['number'].sum().reset_index()
```

```python
plt.figure(figsize=(16,9))
sns.barplot(x='month',y='number',data=data6,dodge=False,hue='month')
plt.legend(bbox_to_anchor=(1.05,1),loc=2)
plt.xticks(rotation=75)
plt.show
```

**Find Average no of fires were reported from highest to lowest (state-wise)**

```python
data7=data.groupby('state')['number'].mean().reset_index()
data7
```

```python
plt.figure(figsize=(16,9))
sns.barplot(x='state',y='number',data=data7,dodge=False,hue='state')
plt.legend(bbox_to_anchor=(1.05,1),loc=2)
plt.xticks(rotation=75)
plt.show
```
